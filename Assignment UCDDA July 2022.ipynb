```python
import pandas as pd
```

```python
df=pd.read_csv("Life Expectancy Data.csv")
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.dropna()
df
```

```python
print (df.columns.tolist())
```

```python
newdf=df[['Year','Country', 'Life expectancy ']]
newdf
```

```python
life_expectancy_2014=newdf.loc[newdf['Year'] == 2014]
life_expectancy_2014
```

```python
life_expectancy_2014=life_expectancy_2014.rename(columns={"Countries":"Country", "Life expectancy ":"Life Expectancy"})
life_expectancy_2014
```

```python
list(life_expectancy_2014.columns)
```

```python
df1 = life_expectancy_2014.sort_values('Life Expectancy',ascending = False).groupby('Country').head(2)
df1
```

```python
top_20_life_expectancy_2014=df1.head(n=20)
top_20_life_expectancy_2014
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (20,20)
```

```python
plt=plt.plot(top_20_life_expectancy_2014["Country"], top_20_life_expectancy_2014["Life Expectancy"])
```

```python
import seaborn as sns
```

```python
sns.barplot(data=top_20_life_expectancy_2014, x='Life Expectancy', y='Country');
```

```python
another_df=pd.read_csv('MIP12.20220716224054.csv')
another_df
```

```python
another_df.isna().sum()
```

```python
another_df.duplicated().sum()
```

```python
updated_df=another_df[['Year','Countries', 'VALUE']]
updated_df
```

```python
life_expectancy_2019=updated_df.loc[updated_df['Year'] == 2019]
life_expectancy_2019
```

```python
life_expectancy_2019=life_expectancy_2019.rename(columns={"Countries":"Country", "VALUE":"Life Expectancy"})
life_expectancy_2019
```

```python
life_expectancy_2019.isna().sum()
```

```python
life_expectancy_2019.duplicated().sum()
```

```python
life_expectancy_2019=life_expectancy_2019.drop_duplicates()
life_expectancy_2019
```

```python
df2 = life_expectancy_2019.sort_values('Life Expectancy',ascending = False).groupby('Country').head(2)
df2
```

```python
top_20_life_expectancy_2019=df2.head(n=21)
top_20_life_expectancy_2019
```

```python
top_20_life_expectancy_2019=top_20_life_expectancy_2019.iloc[1:]
top_20_life_expectancy_2019
```

```python
list(top_20_life_expectancy_2019.columns)
```

```python
top_20_life_expectancy_2019.info()
```

```python
sns.barplot(data=top_20_life_expectancy_2019, x='Life Expectancy', y='Country');
```

```python
combined_data=pd.concat([top_20_life_expectancy_2014, top_20_life_expectancy_2019], ignore_index=True)
combined_data
```

```python
sns.barplot(data=combined_data, x="Life Expectancy", y="Country", hue="Year");
```
