```python
newdf=df[['Year','Country', 'Life expectancy ']]
newdf
# selecting the specific columns I want to work with and the order I want them displayed
```

```python
life_expectancy_2014=newdf.loc[newdf['Year'] == 2014]
life_expectancy_2014
# extracting all rows with the year listed as "2014"
```

```python
life_expectancy_2014=life_expectancy_2014.rename(columns={"Countries":"Country", "Life expectancy ":"Life Expectancy"})
life_expectancy_2014
# renaming columns for merging and to remove trailing values
```

```python
list(life_expectancy_2014.columns)
# checking column names, again for any trailing values etc in case I want to make changes later
```

```python
df1 = life_expectancy_2014.sort_values('Life Expectancy',ascending = False).groupby('Country').head(2)
df1
# sorting rows in descending order by "Life Expectancy"
```

```python
top_20_life_expectancy_2014=df1.head(n=20)
top_20_life_expectancy_2014
# selecting the top 20 rows and assigning new variable name
```

```python
import matplotlib.pyplot as plt
# importing new programme to visualise the data
```

```python
plt.rcParams["figure.figsize"] = (20,20)
# increasing graph size for better visual
```

```python
plt=plt.plot(top_20_life_expectancy_2014["Country"], top_20_life_expectancy_2014["Life Expectancy"])
# plotting the graph
```

```python
import seaborn as sns
# importing seaborn to try achieve better visual
```
